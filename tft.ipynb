```python
from selenium import webdriver
from bs4 import BeautifulSoup
import re
import time

excutable_path = "chromedriver.exe"

url = "https://lolchess.gg/champions/set6/malzahar"
driver = webdriver.Chrome(executable_path=excutable_path)

driver.get(url)
time.sleep(5)
driver.find_element_by_xpath('//*[@id="guide-champion"]/div/table/thead/tr/th[2]').click()
time.sleep(1)
driver.find_element_by_xpath('//*[@id="guide-champion"]/div/table/thead/tr/th[2]').click()
time.sleep(1)
element = driver.find_element_by_xpath('//*[@id="guide-champion"]/div/table').text

print(element)
```

```python
import pandas as pd

champion = element.replace('\n',' ')
champion_list = champion.split(' ')
columns=champion_list[:13]
columns_need=['챔피언','비용','체력','방어력','공격력','공격속도','DPS','마나']
df_columns=pd.DataFrame(columns_need).T
df_columns
```

```python
champion_collect = champion_list[13:]
champion_collect.remove('페이트')
champion_collect.remove('포츈')
champion_collect.remove('박사')
champion_collect.remove('탐')
champion_collect[88]='트위스티드 페이트'
champion_collect[240]='미스포츈'
champion_collect[320]='문도박사'
champion_collect[464]='탐 켄치'
```

```python
champion_collect
```

```python
champion_collection= []

for i in range(59):
    champion_collection.append(champion_collect[i*8:(i*8)+8])

champion_collection
```

```python
df_columns=pd.DataFrame(champion_collection,columns=columns_need)
df_champions = df_columns.set_index('챔피언')
```

```python
temp_img_alt = driver.find_element_by_xpath('//*[@id="guide-champion"]/div/table/tbody/tr[1]/td[3]/div/div/img')
print(temp_img_alt.get_attribute('alt'))
```

```python
class_list = []
for i in range(59):
    img = driver.find_element_by_xpath('//*[@id="guide-champion"]/div/table/tbody/tr[{}]/td[3]/div/div/img'.format(i+1))
    class_list.append(img.get_attribute('alt'))
class_list
```

```python
pd.DataFrame(class_list, columns=["계층"])
```

```python
job_list = []
for i in range(51):
    img = driver.find_element_by_xpath('//*[@id="guide-champion"]/div/table/tbody/tr[{}]/td[4]/div/div/img'.format(i+1))
    job_list.append(img.get_attribute('alt'))
job_list
```

```python
job_list.append(' ')
len(job_list)
```

```python
url = "https://lolchess.gg/champions/set6/chogath"
driver = webdriver.Chrome(executable_path=excutable_path)
driver.get(url)
time.sleep(1)
driver.find_element_by_xpath('//*[@id="guide-champion"]/div/table/thead/tr/th[2]').click()
time.sleep(1)
driver.find_element_by_xpath('//*[@id="guide-champion"]/div/table/thead/tr/th[2]').click()
time.sleep(1)
item_list=[]
for i in range(1,60):
    driver.find_element_by_xpath('//*[@id="guide-champion"]/div/table/tbody/tr[{}]/td[1]/a[2]/span'.format(i)).click()
    time.sleep(1)
    for j in range(1,6):
        url = driver.find_element_by_xpath('//*[@id="wrapper"]/div/div[6]/div/div[3]/div/div[6]/div[2]/div[{}]'.format(j)).get_attribute('data-tooltip-url')
        driver.get(url)
        time.sleep(1)
        item = driver.find_element_by_xpath('/html/body/div/p')
        item_list.append(item.text)
        driver.back()
        time.sleep(1)
    driver.find_element_by_xpath('//*[@id="guide-champion"]/div/table/thead/tr/th[2]').click()
    time.sleep(1)
    driver.find_element_by_xpath('//*[@id="guide-champion"]/div/table/thead/tr/th[2]').click()
    time.sleep(1)
```

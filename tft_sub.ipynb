# 레벨 증강에 따른 공격력 수치

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import re
import time

excutable_path = "chromedriver.exe"

url = "https://lolchess.gg/champions/set6/talon"
driver = webdriver.Chrome(executable_path=excutable_path)
driver.get(url)
time.sleep(1)
driver.find_element_by_xpath('//*[@id="guide-champion"]/div/table/thead/tr/th[2]').click()
time.sleep(1)
driver.find_element_by_xpath('//*[@id="guide-champion"]/div/table/thead/tr/th[2]').click()
time.sleep(1)
skill_information=[]
for i in range(1,60):
    driver.find_element_by_xpath('//*[@id="guide-champion"]/div/table/tbody/tr[{}]/td[1]/a[2]/span'.format(i)).click()
    time.sleep(1)
    second = driver.find_element_by_xpath('//*[@id="wrapper"]/div/div[6]/div/div[3]/div/div[3]/div[2]/div[2]/div')
    skill_information.append(second.text)
    time.sleep(1)
    driver.find_element_by_xpath('//*[@id="guide-champion"]/div/table/thead/tr/th[2]').click()
    time.sleep(1)
    driver.find_element_by_xpath('//*[@id="guide-champion"]/div/table/thead/tr/th[2]').click()
    time.sleep(1)
```

```python
skill_information
```

```python
skill_information_spl= []
for i in skill_information:
    skill_information_spl.append(i.split('/'))
```

```python
import pandas as pd

skill_df = pd.DataFrame(skill_information_spl,columns=["2성 공격력 ","3성 공격력"])
skill_df
```

```python
skill_df.to_excel(excel_writer='skill_information.xlsx')
```

# 마나 부분 재전처리

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import re
import time

excutable_path = "chromedriver.exe"

url = "https://lolchess.gg/champions/set6/orianna"
driver = webdriver.Chrome(executable_path=excutable_path)
driver.get(url)
time.sleep(1)
driver.find_element_by_xpath('//*[@id="guide-champion"]/div/table/thead/tr/th[2]').click()
time.sleep(1)
driver.find_element_by_xpath('//*[@id="guide-champion"]/div/table/thead/tr/th[2]').click()
time.sleep(1)
mana_information=[]
for i in range(1,60):
    mana=driver.find_element_by_xpath('//*[@id="guide-champion"]/div/table/tbody/tr[{}]/td[12]'.format(i))
    time.sleep(1)
    mana_information.append(mana.text)
print(mana_information)
```

```python
mana_information_spl= []
for i in mana_information:
    mana_information_spl.append(i.split('/'))
print(mana_information_spl)
```

```python
mana_df = pd.DataFrame(mana_information_spl,columns=["보유마나","최대마나"])
mana_df
```

```python
mana_df.to_excel(excel_writer='mana_information.xlsx')
```
